## 00 Setup

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the file path
file_path = 'data/stadium.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df
```

```python
# Checking the structure of the data
df.info()
df.describe()
```

## 01 Stadium Capacity Analysis:

1. **What is the distribution of stadium capacities by confederation?**

   This question is asking about the arrangement or spread of stadium capacities (the number of seats or spectators a stadium can accommodate) across different confederations. In the context of international football, a confederation refers to a regional governing body that oversees football (soccer) in a specific part of the world, such as UEFA for Europe, CONCACAF for North and Central America, and so on.

   To answer this question, you would typically:
   - Group the stadiums in your dataset by their confederation (e.g., UEFA, CONCACAF, AFC).
   - Calculate summary statistics or create visualizations (e.g., histograms, box plots, or bar charts) to show how stadium capacities are distributed within each confederation.
   - This analysis helps identify the typical stadium sizes in different regions and whether there are any outliers (stadiums with very high or very low capacities) in specific confederations.

2. **Are there significant variations in stadium sizes within each confederation?**

   This question is inquiring about the level of diversity or differences in stadium capacities within each confederation. It's about understanding whether, within a given region or confederation, there is a wide range of stadium sizes or if most stadiums have similar capacities.

   To address this question, you would typically:
   - Focus on each confederation separately (e.g., UEFA, CONCACAF, AFC) and examine the stadium capacities within that specific region.
   - Calculate measures of spread or dispersion (e.g., standard deviation or interquartile range) to quantify the degree of variation in stadium capacities within each confederation.
   - Create visual representations, such as box plots or histograms, to illustrate the distribution of stadium capacities in each confederation.
   - Identify if there are regions where stadiums tend to have capacities that are close to each other or if there is a significant variation, with some stadiums being much larger or smaller than the average for that confederation.

```python
# Group the data by confederation and calculate the average capacity in each confederation
confederation_capacity = df.groupby('Confederation')['Capacity'].mean().sort_values(ascending=False)

# Create a bar plot to visualize the distribution
plt.figure(figsize=(12, 6))
confederation_capacity.plot(kind='bar')
plt.title('Average Stadium Capacity by Confederation')
plt.xlabel('Confederation')
plt.ylabel('Average Capacity')
plt.xticks(rotation=45)
plt.show()
```

## 02 Country Comparison:

1. **Which countries have the highest and lowest average stadium capacities?**

   - **Average Stadium Capacity**: This part of the question focuses on calculating the average (mean) stadium capacity for each country in the dataset. This means aggregating the stadium data based on the countries they are located in and finding the average capacity for each country.

   - **Highest and Lowest**: The goal is to identify the countries with the highest and lowest average stadium capacities. This information can provide insights into which countries, on average, have the largest and smallest stadiums.

2. **Is there a correlation between stadium capacity and a country's population or economic development?**

   - **Correlation Analysis**: This part of the question delves into a deeper analysis. It seeks to determine whether there is a statistical relationship or correlation between stadium capacity and certain country-level factors: population and economic development.

   - **Population**: By examining the correlation between stadium capacity and a country's population, you aim to understand if countries with larger populations tend to have stadiums with higher capacities. This can help answer questions about how stadiums are distributed relative to the size of the population.

   - **Economic Development**: If you have data on a country's economic development, such as GDP, this part of the question explores whether countries with higher economic development tend to have larger stadiums. It can provide insights into the connection between economic prosperity and the development of sports infrastructure.

```python
# Group the data by country and calculate the average stadium capacity and population
average_capacity_by_country = df.groupby('Country')['Capacity'].mean().sort_values(ascending=False)
country_population = df.groupby('Country')['Population'].mean()

# Create a bar plot to visualize countries with the highest average stadium capacities
plt.figure(figsize=(12, 6))
average_capacity_by_country.head(10).plot(kind='bar', color='green', alpha=0.7, label='Top 10')
plt.title('Countries with Highest Average Stadium Capacities')
plt.xlabel('Country')
plt.ylabel('Average Capacity')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a bar plot to visualize countries with the lowest average stadium capacities
average_capacity_by_country.tail(10).plot(kind='bar', color='red', alpha=0.7, label='Bottom 10')
plt.title('Countries with Lowest Average Stadium Capacities')
plt.xlabel('Country')
plt.ylabel('Average Capacity')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatter plot to explore the correlation between stadium capacity and population
plt.figure(figsize=(12, 6))
plt.scatter(country_population, average_capacity_by_country, alpha=0.5)
plt.title('Correlation between Stadium Capacity and Population')
plt.xlabel('Population')
plt.ylabel('Average Capacity')
plt.show()
```

```python
# Sort the data by population for a smooth line plot
sorted_population = country_population.sort_values()
sorted_capacity = average_capacity_by_country.loc[sorted_population.index]

# Create a line plot to explore the correlation
plt.figure(figsize=(12, 6))
plt.plot(sorted_population, sorted_capacity, marker='o', linestyle='-')
plt.title('Correlation between Stadium Capacity and Population')
plt.xlabel('Population')
plt.ylabel('Average Capacity')
plt.grid(True)
plt.show()
```

## 03 Stadiums by Confederation:

1. **Which confederation has the most stadiums, and what is the average capacity within each confederation?**

   - This question seeks to identify the confederation with the highest number of stadiums. A confederation is an international organization that governs football in a specific region, such as UEFA for Europe or CONCACAF for North and Central America. The answer to this part of the question will provide insight into which region has the most football stadiums.
   
   - Additionally, the question also asks for the average stadium capacity within each confederation. This means you'll calculate the average size of stadiums for each confederation. The average capacity is a measure of how large, on average, the stadiums are within each region.

   **Why is this question important?** This analysis can help assess the regional distribution of stadiums in international football. It can reveal which confederation is likely to have a higher concentration of football facilities and provide information about the size of stadiums in each region, which can be valuable for various purposes, such as event planning or understanding regional football culture.

2. **Are there any trends or differences in stadium distribution by confederation?**

   - This question is interested in understanding if there are any patterns, trends, or differences in the distribution of stadiums across different confederations. It's looking for variations in stadium count and size between confederations.
   
   **Why is this question important?** Identifying trends or differences in stadium distribution can provide valuable insights into the global football landscape. It may reveal regions where football infrastructure is more developed or regions where stadium sizes differ significantly, possibly due to cultural, economic, or historical factors. Understanding these trends can inform strategic decisions and policies related to football.

```python
# Group the data by confederation and calculate the number of stadiums and average capacity in each confederation
confederation_stats = df.groupby('Confederation').agg({'Stadium': 'count', 'Capacity': 'mean'})

# Sort confederations by the number of stadiums in descending order
confederation_stats = confederation_stats.sort_values(by='Stadium', ascending=False)

# Create a bar plot to visualize the number of stadiums in each confederation
plt.figure(figsize=(12, 6))
confederation_stats['Stadium'].plot(kind='bar', color='blue')
plt.title('Number of Stadiums by Confederation')
plt.xlabel('Confederation')
plt.ylabel('Number of Stadiums')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group the data by confederation and calculate the number of stadiums
confederation_counts = df['Confederation'].value_counts()

# Create a pie chart to visualize the distribution of stadiums by confederation
plt.figure(figsize=(8, 8))
plt.pie(confederation_counts, labels=confederation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Stadiums by Confederation')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Create a bar plot to visualize the average capacity within each confederation
plt.figure(figsize=(12, 6))
confederation_stats['Capacity'].plot(kind='bar', color='green')
plt.title('Average Capacity by Confederation')
plt.xlabel('Confederation')
plt.ylabel('Average Capacity')
plt.xticks(rotation=45)
plt.show()
```

## 04 City Stadiums Analysis:

1. **Which cities have the most football stadiums, and what is the average capacity in those cities?**

   This question aims to identify and rank cities based on the number of football stadiums they host. It also seeks to determine the average stadium capacity in those cities. By doing this, you can pinpoint cities that are particularly well-known for having a high concentration of football stadiums and understand the typical size of stadiums in these cities. 

   **Why is this question important?** This analysis can be valuable for sports enthusiasts, event organizers, and city planners. It helps in understanding where stadiums are most concentrated and their typical sizes, which may influence event planning, infrastructure development, and sports-related decisions.

2. **Are there cities with a concentration of stadiums?**

   This question delves further into the dataset to identify cities that exhibit a concentration of football stadiums, specifically those with multiple stadiums in close proximity to each other. By determining a threshold (e.g., a certain number of stadiums), you can find cities with a significant number of stadiums clustered together.

   **Why is this question important?** Identifying cities with a concentration of stadiums can be of interest to event organizers, tourism boards, and urban planners. It may indicate areas where sports events and fan engagement are highly concentrated, which could inform marketing strategies and urban development plans.

```python
# Group the data by city and calculate the number of stadiums and the average capacity in each city
city_stadium_counts = df['City'].value_counts()
city_avg_capacity = df.groupby('City')['Capacity'].mean()

# Determine the cities with the most stadiums
most_stadium_cities = city_stadium_counts[city_stadium_counts == city_stadium_counts.max()]

# Create a DataFrame for cities with the most stadiums and their average capacities
most_stadium_df = pd.DataFrame({
    # 'City': most_stadium_cities.index,
    'Number of Stadiums': most_stadium_cities.values,
    'Average Capacity': city_avg_capacity.loc[most_stadium_cities.index]
})

# Identify cities with a concentration of stadiums
concentration_threshold = 5  # You can adjust this threshold as needed
concentration_cities = city_stadium_counts[city_stadium_counts >= concentration_threshold]

# Create a DataFrame for cities with a concentration of stadiums and their average capacities
concentration_df = pd.DataFrame({
    # 'City': concentration_cities.index,
    'Number of Stadiums': concentration_cities.values,
    'Average Capacity': city_avg_capacity.loc[concentration_cities.index]
})

# Display the DataFrames in a tabular format
print("Cities with the most football stadiums and their average capacities:")
print(most_stadium_df)

print("\nCities with a concentration of stadiums ({} or more stadiums) and their average capacities:".format(concentration_threshold))
print(concentration_df)
```
